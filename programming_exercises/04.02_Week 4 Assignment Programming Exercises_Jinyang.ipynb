1. What is the average total yearly earnings across all of Boston?

```python
import pandas
import matplotlib.pyplot as plt
df = pandas.read_csv("../data/boston_earning_data.csv", sep=',', encoding='utf8')
df.head()
```

```python
df["TOTAL_PAY"].mean()
```

2. What is the average in overtime pay across Boston?

```python
df.columns
df["OVERTIME_PAY"].mean()
```

3. Which three departments have the highest average earnings, which three have the lowest?

```python
df_grouped=df.groupby("DEPARTMENT_NAME")
df_grouped_mean=df_grouped.mean()
type(df_grouped_mean["TOTAL_PAY"])
df_grouped_mean["TOTAL_PAY"]
```

```python
print(df_grouped_mean["TOTAL_PAY"].idxmax())
index_min_1=df_grouped_mean["TOTAL_PAY"].idxmin()
index_min_2=df_grouped_mean["TOTAL_PAY"].drop(index_min_1).idxmin()
index_min_3=df_grouped_mean["TOTAL_PAY"].drop([index_min_1,index_min_2]).idxmin()
print([index_min_1,index_min_2, index_min_3])
#can use mean.sort_values(ascending=False)[:3]
```

4. Produce a bar chart that visualizes the average earning by department.

```python
totalpay_mean=df_grouped_mean["TOTAL_PAY"]
totalpay_mean.plot(kind='barh')
plt.legend(loc=0, bbox_to_anchor=(1.3, 0.5))
plt.show()
```

5. Produce a bar chart the visualizes the average overtime earning by department.

```python
overtimepay_mean=df_grouped_mean["OVERTIME_PAY"]
overtimepay_mean.plot(kind='barh')
plt.legend(loc=0, bbox_to_anchor=(1.35, 0.5))
plt.show()
```

6. What is the difference in average total earning for the department that earns the most, and the department that earns the least, on average? (An approximation of inequality)

```python
df_grouped_mean["TOTAL_PAY"].max()-df_grouped_mean["TOTAL_PAY"].min()
```
