**3. Create a dataframe including articles that contain keyword 'China' from 2000 to 2022**

```python
start_root = "0101"
end_root = "1231"
q = ''
df = pd.DataFrame()

for date in range(2000, 2023):
    try:
        start_date = str(date)+start_root
        end_date = str(date)+end_root
        print(start_date)
        search_params = {"q": "China",
                     "api-key": key,
                    "begin_date": start_date,
                    "end_date": end_date}
        r = requests.get(base_url, params=search_params)
        response_text = r.text
        data = json.loads(response_text)
        data_response = data['response']['docs']
        df1=pd.DataFrame(data_response)
        print(df1)
        df=pd.concat([df, df1])

        time.sleep(5)
    except:
        print(r.url)
        print("Oh no, something went wrong.")
df
```

**4. Add a 'year' column showing the year of each article**

```python
df['year'] = df['web_url'].str.extract(r'(\d{4})')
df
```

**5. The average word count per article for each year**

```python
df_grouped=df.groupby("year")
df_grouped.mean()
```

```python
df_mean=df_grouped.mean()
df_cn_mean=df_mean.drop(index='1247')
df_cn_mean
```

**6. Create a dataframe including articles that contain keyword 'Japan' from 2000 to 2022**

```python
start_root = "0101"
end_root = "1231"
q = ''
df_japan = pd.DataFrame()

for date in range(2000, 2023):
    try:
        start_date = str(date)+start_root
        end_date = str(date)+end_root
        print(start_date)
        search_params = {"q": "Japan",
                     "api-key": key,
                    "begin_date": start_date,
                    "end_date": end_date}
        r = requests.get(base_url, params=search_params)
        response_text = r.text
        data = json.loads(response_text)
        data_response = data['response']['docs']
        df1=pd.DataFrame(data_response)
        print(df1)
        df_japan=pd.concat([df_japan, df1])

        time.sleep(5)
    except:
        print(r.url)
        print("Oh no, something went wrong.")
df_japan
```

```python
df_japan['year'] = df_japan['web_url'].str.extract(r'(\d{4})')
df_japan
df_jp_grouped=df_japan.groupby("year")
df_jp_grouped.mean()
```

```python
df_j_mean=df_jp_grouped.mean()
df_j_mean.drop(index='1000')
df_jp_mean=df_j_mean.drop(index='1000')
df_jp_mean
df_jp_mean.columns
```

**7. Convert the year from 4 digit to 2 digit**

```python
df_cn_mean.index=pd.to_datetime(df_cn_mean.index, format='%Y').strftime('%y')
df_jp_mean.index=pd.to_datetime(df_jp_mean.index, format='%Y').strftime('%y')
```

**8. Visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14,6))

sns.lineplot(y='word_count', x=df_cn_mean.index, data=df_cn_mean, color = 'red')
sns.lineplot(y='word_count', x=df_jp_mean.index, data=df_jp_mean, marker="o")

ax.set_facecolor('white') #set background as white

#remove spines around the chart
ax.spines['top'].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

#label the lines
for line, name in zip(ax.lines, ['China', 'Japan']):
    y = line.get_ydata()[-1]
    ax.annotate(name, xy=(0.96,y), xytext=(6,0), 
                color=line.get_color(), 
                xycoords = ax.get_yaxis_transform(),
                textcoords="offset points",
                size=14, 
                va="center", 
                family='serif')

plt.xlabel('') #year is obvious, we don't need the label
plt.xticks(fontsize=14, family='serif') #set font and size of year on x-axis

plt.ylabel('mean of word counts', fontsize=14, family='serif') #set y-label
plt.yticks(fontsize=12, family='serif') #set font and size of year on x-axis

#control ticks
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=True) # labels along the bottom edge are on

plt.tick_params(
    axis='y',          # changes apply to the y-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the left edge are off
    labelbottom=True) # labels along the bottom edge are on

plt.tight_layout() #make sure the layout is efficient
```

The graph shows that generally, from 2000 to 2021, New York Times pays more attention to China than Japan (using the average word counts each year as an index). Only in 2011 and 2015-2017, attention to Japan has surpassed attention to China. Attention to China has two peaks: 2005 and 2018.
